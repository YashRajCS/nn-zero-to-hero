```python
# Info:
# X -> (n_samples, n_features) - we will have 569 samples, 30 features
# Y -> (n_samples, ) - a flat vector of 0s and 1s (malignant, benign)
# w -> (n_features, ) - one weight per feature
# b -> a single scalar
```

```python
# imports
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# load and split the data
data = load_breast_cancer()
X, Y = data.data, data.target # X: inputs, Y: labels

# splitting into train, test (80%, 20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # fixing a seed for reproducability
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
# standardizing X
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

print(X_train.shape, X_test.shape)
```

```python
# parameters (W, b)
np.random.seed(42)
W = np.random.randn(30)
b = 0.0
```

```python
def sigmoid(z):
    return (1 + np.exp(-z))**(-1)

for i in range(300000):
    # forward pass
    y_pred = sigmoid(X_train @ W + b)
    loss = -np.mean(Y_train * np.log(y_pred) + (1 - Y_train) * np.log(1 - y_pred))

    # if i % 100 == 0: print(f'step {i}: loss {loss}')
    
    # gradients
    dw = X_train.T @ (y_pred - Y_train) / len(Y_train)
    db = np.mean(y_pred - Y_train)
    
    # updates
    lr = 0.01
    W += -lr * dw
    b += -lr * db

print(loss)
```

```python
# evaluate on the test set
y_pred_test = sigmoid(X_test @ W + b)
test_loss = -np.mean(Y_test * np.log(y_pred_test) + (1 - Y_test) * np.log(1 - y_pred_test))

# test accuracy
preds = (y_pred_test > 0.5).astype(int) # get 0 or 1 to compare with label
accuracy = np.mean(preds == Y_test)

print(f"test loss:     {test_loss:.4f}")
print(f"test accuracy: {accuracy:.4f}")
```
