```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read all the words
words = open('names.txt', 'r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
# build the vocabulary of characters and mappings to/from integers
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
print(itos)
```

```python
# build the dataset

block_size = 3 # context length: how many characters do we take to predit the next one?
X, Y = [], [] # X is input to NN, Y are the labels for each example in X
for w in words:
    
    # print(w)
    context = [0] * block_size
    
    for ch in w + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        # print(''.join(itos[i] for i in context), '--->', itos[ix])
        context = context[1:] + [ix] # crop and append

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
# build the dataset

def build_dataset(words):
    block_size = 3 # context length: how many characters do we take to predit the next one?
    X, Y = [], [] # X is input to NN, Y are the labels for each example in X
    for w in words:
        
        # print(w)
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            # print(''.join(itos[i] for i in context), '--->', itos[ix])
            context = context[1:] + [ix] # crop and append
    
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27, 10), generator=g)
W1 = torch.randn((30, 200), generator=g) # 3 inputs x 2 dimensions (from prev line) = 6, and 100 neurons (hyperparameter - up to us to decide)
b1 = torch.randn(200, generator=g)
W2 = torch.randn((200, 27), generator=g) # 100 inputs, and 27 neurons (27 possible characters that come next)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
sum(p.nelement() for p in parameters) # number of parameters in total
```

```python
for p in parameters:
    p.requires_grad = True
```

```python
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre
```

```python
lri = []
lossi = []
stepi = []
```

```python
for i in range(200000):

    # minibatch construct
    ix = torch.randint(0, Xtr.shape[0], (32,))
    
    # forward pass
    emb = C[Xtr[ix]] # 32 x 3 x 2 = 96 x 2
    h = torch.tanh(emb.view(-1, 30) @ W1 + b1)
    logits = h @ W2 + b2
    # counts = logits.exp()
    # prob = counts / counts.sum(1, keepdims=True)
    # loss = -prob[torch.arange(32), Y].log().mean() # negative log-likelihood loss function
    # done the same way using:
    loss = F.cross_entropy(logits, Ytr[ix]) # helps with efficiency and better ops
    # print(loss.item())
    
    # backward pass
    for p in parameters: # reset
        p.grad = None
    loss.backward()
    
    # update
    # lr = lrs[i]
    lr = 0.1 if i < 100000 else 0.01
    for p in parameters:
        p.data += -lr * p.grad

    # track stats
    # lri.append(lre[i])
    stepi.append(i)
    lossi.append(loss.log10().item())

print(loss.item())
```

```python
# training split, dev/validation split, test split
# 80%, 10%, 10%
```

```python
# sample from the model
g = torch.Generator().manual_seed(2147483647)

for _ in range(20):

    out = []
    context = [0] * block_size # initialize with all ...
    while True:
        emb = C[torch.tensor([context])]
        h = torch.tanh(emb.view(1, -1) @ W1 + b1)
        logits = h @ W2 + b2
        probs = F.softmax(logits, dim=1)
        ix = torch.multinomial(probs, num_samples=1, generator=g).item()
        context = context[1:] + [ix]
        out.append(ix)
        if ix == 0: break

    print(''.join(itos[i] for i in out))
```
